```python
diag_1 = diag.ID.str.split(expand = True)
```

```python
diag_1
```

```python
for i in range(len(diag_1.columns)):
    diag_1.iloc[:,i] = diag_1.iloc[:,i].str.strip('[')
    diag_1.iloc[:,i] = diag_1.iloc[:,i].str.strip(']')
    diag_1.iloc[:,i] = diag_1.iloc[:,i].str.strip(',')
    diag_1.iloc[:,i] = diag_1.iloc[:,i].str.strip("''")
```

```python
diag_1
```

```python
diag_1['ID'] = diag_1['ID'].astype(int)
```

```python
diag_2 = pd.melt(frame=diag_1, id_vars='ID')
```

```python
diag_2 = diag_2.drop(['variable'], axis='columns')
```
